```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/Ezeysay16/veridosyalari/main/Fifa_world_cup_matches.csv"
data = pd.read_csv(url)
data.head()
```

**Soru 1**

*The worst game in the world cup is the one where a team
  dominates too much the other teams (i.e posession ratio is the highest)
  Implement this function such that it finds the highest ratio between 
  'possession team1' and 'possession team2' (and its inverse) to find the worst 
  game and returns the following info for that game:
    team1, team2, date, possession team1 and possession team2*

```python
data["possession team1"] = data["possession team1"].str.rstrip("%").astype(int)
data["possession team2"] = data["possession team2"].str.rstrip("%").astype(int)
data["dif"] = abs(data["possession team1"] - data["possession team2"])
```

```python

data.loc[data["dif"].argmax()]
```

```python
team1 = data.loc[data["dif"].argmax(), "team1"]
team2 = data.loc[data["dif"].argmax(), "team2"]
date = data.loc[data["dif"].argmax(), "date"]
pos1 = data.loc[data["dif"].argmax(), "possession team1"]
pos2 = data.loc[data["dif"].argmax(), "possession team2"]
```

```python
print(team1, 'vs',team2, date, 'possession team1:', pos1, 'possession team2:', pos2)
```

**Soru 2**

*Efficiency of a team is defined by the number of goals divided by possession.
  Implement this function such that it returns the name of the most efficient 
  team and the value for efficiency (number of goals by possesion). Take into
  account that a team may appear as team1 or team2 in a game.*

```python
data.head()
```

```python
data["team1_eff"] = data["number of goals team1"] / data["possession team1"]
data["team2_eff"] =  data["number of goals team2"] / data["possession team2"]
data.head()
max_eff = 0
most_eff_team = "" 
for team in data["team1"].unique():
    eff = (data[data["team1"] == team]["team1_eff"].mean() + 
                    data[data["team2"] == team]["team2_eff"].mean()) / 2
    if eff > max_eff:
        max_eff = eff
        most_eff_team = team

        
```

```python
print('----MOST EFFICIENT TEAM ----')
print(most_eff_team, 'efficiency:', max_eff)
```
